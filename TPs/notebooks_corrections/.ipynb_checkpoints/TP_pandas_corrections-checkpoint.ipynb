### Question 1.2
Here is a list of data about planets in our solar system. The mass and distance are given relative to the Earth. Create a `Series` for the mass of the planets. The index should be the planets' names.

### Question 2.1

### Question 2.2
Sort `df_planet` by the mass in descending order to find the heaviest planets

```python
df_planet.sort_values('mass_kg', ascending=False)
```

### Question 2.3
Here is a new series called `type_of_planets` which contains the type of each of the planets.
Add a new column to `df_planet` containing the information of `type_of_planets`.

```python
type_of_planets = pd.Series(['dwarf','dwarf','gaz','dwarf',
                             'gaz', 'dwarf','dwarf','gaz',
                             'terrestrial','terrestrial', 
                             'gaz','terrestrial','terrestrial'],
                             index = df_planet.index)
type_of_planets
```

```python
# write your answer here ...
df_planet['type_of_planet'] = type_of_planets
```

### Question 2.4
Group the planets by their type and compute the mean density.

```python
df_planet.groupby('type_of_planet').meanRadius.mean()
```

### Question 2.5
Now using the same group and the `agg` method, compute the mean, minimum, maximum and standard deviation of the mean radius.

```python
df_planet[['meanRadius','type_of_planet']]\
        .groupby('type_of_planet')\
        .agg(['mean','max','min','var']).meanRadius
```

### <span style="color:red">Question 2.6:</span>
<span style="color:red">*More open question*</span>


The column *orbit* specifies around which other body a given body is orbitting. Find which planet has the most satellites.

```python
all_orbits = df.groupby(df.orbits).count().isPlanet
planets_with_satellites = [planet for planet in df_planet.index if planet in all_orbits.index]
orbits_planets = all_orbits.loc[planets_with_satellites]
sorted_orbits_planets = orbits_planets.sort_values(ascending=False)
print(f"The planet with the most satellites is {sorted_orbits_planets.index[0]} with {sorted_orbits_planets.values[0]} satellites.")
```

## Question 4.1

Open the following files
- `../data/PerosBanhos-NorthPB-IleDiamant_71.76952_-5.24626_10_.csv`
- `../data/PerosBanhos-NorthPB-IleDiamant_71.76952_-5.24626_25_.csv`

**Attention !** Only use the first 300 000 rows as files are very large.

Use the keyword `parse_dates=True` to automatically convert dates into `datetime_index`

```python
# write your answer here
df10 = pd.read_csv('../data/PerosBanhos-NorthPB-IleDiamant_71.76952_-5.24626_10_.csv', 
                   index_col=0, nrows=300000, parse_dates=True)
df25 = pd.read_csv('../data/PerosBanhos-NorthPB-IleDiamant_71.76952_-5.24626_25_.csv', 
                   index_col=0, nrows=300000, parse_dates=True)
```

### Question 4.2

Combine the two timeseries into a unique dataframe, where the variables are ``T10m`` and ``T25m``

```python
# write your answer here
# option 1
df = pd.DataFrame({'T25m':df25.Temperature, 'T10m': df10.Temperature}, index=df10.index)
# option 2
df = pd.concat([df25, df10], axis=1)
df.columns = ["T25m", "T10m"]

df
```

### Question 4.3

Plot the two timeseries on April 24th 2019 as line plots.

```python
# write your answer here

df.loc['2019-04-24'].plot()
```

### Question 4.4
Get the key statistics of the dataframe `df` using the describe method.

```python
# write your answer here

df.describe()
```

### Question 4.5 
Resample the data to get one point every 10 minutes, and plot the same day.

You can try to use a mean aggregator or a minimum.

```python
# write your answer here

df_10min = df.resample('10min').min()
df_10min.loc['2019-04-24'].plot()
```

### Question 4.6
Plot the data on the day where there is the largest temperature difference between 10m and 25m.
To do so, you could use the `idxmax` method

```python
temp_difference = df.T10m - df.T25m
day_selected = temp_difference.idxmax().strftime('%Y-%m-%d')
df_10min.loc[day_selected].plot()
```

### Question 4.7
Using the full dataset, extract data that occur only from 12:00 to 12:59. Plot a histogram of the data

```python
# write your answer here

df_10min.loc[df_10min.index.hour==12].plot.hist(alpha=.5, bins=20)
```

### Question 4.8
Using the 10min resampled dataset, compute the mean diurnal cycle of the data, grouping data by hour.
Plot the results as line plots

```python
# write your answer here

diurnal_cycle = df_10min.groupby(df_10min.index.hour).mean()
diurnal_cycle.plot()
```

### Question 4.9
Using the same 10min resampled dataset, compute the mean diurnal cycle of the data, grouping data by hour and minutes.
Plot the results as line plots

```python
# write your answer here

diurnal_cycle_minutes = df_10min.groupby(df_10min.index.hour*60+df_10min.index.minute).mean()
diurnal_cycle_minutes.index = diurnal_cycle_minutes.index/60
diurnal_cycle_minutes.plot()

diurnal_cycle_minutes_quantiles = df_10min.T25m.groupby(df_10min.index.hour*60+df_10min.index.minute).quantile([0.17,0.83])
diurnal_cycle_minutes_quantiles = diurnal_cycle_minutes_quantiles.unstack()
diurnal_cycle_minutes_quantiles.index = diurnal_cycle_minutes_quantiles.index/60
```

### <span style="color:red">Question 4.10:</span>
<span style="color:red">*More open question*</span>

Compute the quantiles 0.25 and 0.75 on the T25m and on the T10m to compute uncertainties in the diurnal cycle and add it to the plot using the `plt.fill_between(x, y1, y2)` matplotlib function. You will need to use reshaping functions such as `df.unstack()`. 
You can also change the aggregator used for the line from `mean` to `median` for better results.


```python
# write your answer here

diurnal_cycle_minutes = df_10min.groupby(df_10min.index.hour*60+df_10min.index.minute).median()
diurnal_cycle_minutes.index = diurnal_cycle_minutes.index/60
diurnal_cycle_minutes.plot()

diurnal_cycle_minutes_quantiles_T25 = df_10min.T25m.groupby(df_10min.index.hour*60+df_10min.index.minute).quantile([0.25,0.75])
diurnal_cycle_minutes_quantiles_T25 = diurnal_cycle_minutes_quantiles_T25.unstack()
diurnal_cycle_minutes_quantiles_T25.index = diurnal_cycle_minutes_quantiles_T25.index/60

diurnal_cycle_minutes_quantiles_T10 = df_10min.T10m.groupby(df_10min.index.hour*60+df_10min.index.minute).quantile([0.25,0.75])
diurnal_cycle_minutes_quantiles_T10 = diurnal_cycle_minutes_quantiles_T10.unstack()
diurnal_cycle_minutes_quantiles_T10.index = diurnal_cycle_minutes_quantiles_T10.index/60

plt.fill_between(diurnal_cycle_minutes_quantiles_T25.index, 
                 diurnal_cycle_minutes_quantiles_T25[0.25],
                 diurnal_cycle_minutes_quantiles_T25[0.75],
                 alpha = .2
                )             

plt.fill_between(diurnal_cycle_minutes_quantiles_T10.index, 
                 diurnal_cycle_minutes_quantiles_T10[0.25],
                 diurnal_cycle_minutes_quantiles_T10[0.75],
                 alpha = .2
                )             
```
